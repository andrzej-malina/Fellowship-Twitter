```python
# For each row in DataFrame execute INSERT INTO query
def upload_to_db(data_frame, connection, file_name):
    # For each row in DataFrame execute INSERT INTO query
    t0 = time.time()
    rows_with_error = []  # empty list for not uploaded data
    q = """INSERT INTO tweets (Username, Name, Datetime, Tweet_Id, Text, Tweet_Retweet_count, Tweet_Favorite_count, Tweet_language, Source, Location, Company_url, User_followers_count, User_friends_count)
            VALUES (%s, %s , %s, %s ,%s, %s ,%s ,%s ,%s ,%s ,%s ,%s ,%s)"""
    values = [tuple(x) for x in data_frame.values]  # list of tuples with values to upload
    for val in values:
        cursor = connection.cursor()
        try:
            try:
                cursor.execute(q, val)
            except Exception as e:
                connection.rollback()
                rows_with_error.append(val)
                print(e)
                # raise
            else:
                connection.commit()
        finally:
            cursor.close()
    print(file_name, ' Packet upload time %.2f sec' % (time.time() - t0))
    return rows_with_error
```

```python
# Get each of the pickle format data, unpickle Dataframe and run upload_to_DB function, 
#returns list of all the values that were not uploaded
def update_db(files_list, db_connection):
    # Get each of the pickle format data from the list, unpickle Dataframe and run upload_to_db function,
    # returns list of all the values that were not uploaded
    not_uploaded_data = []  # empty list for not uploaded data
    for pkl_file in files_list:
        try:
            with open(FILE_LOCATION + pkl_file, mode='rb') as file:  # open each pickle
                data = pickle.load(file)
            file.close()
            not_uploaded_data.extend(upload_to_db(data, db_connection, pkl_file))
        except:
            print('Cannot open file ' + pkl_file)
            continue
    print('Completed upload')
    return not_uploaded_data
```
